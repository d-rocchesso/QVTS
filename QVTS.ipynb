## Operators

$\sigma$: a vector operator informing about the phonetic elements along a specific "direction" of measurement. 

- $\sigma_z$: Phonation. Eigenstates: upper and lower pitch 
- $\sigma_x$: Turbulence. Eigenstates: leftmost and rightmost centroids of two distributions of noise
- $\sigma_y$: Slow myoelastic vibrations. Eigenstates: faster and slower pulsation tempos

Relation with the sines + noise + transients model ([Verma, Levine, and Meng, 1997](https://ccrma.stanford.edu/files/papers/stanm101.pdf#page=50))

The **Spin Polarization Principle**: there is no sound state for which the expectation value of the three components is zero.

A measurement of turbulence **prepares** the "system" in one of two eigenstates for $\sigma_x$, and a successive measurement of phonation would find a superposition and get equal probabilities for the two eigenstates of
$\sigma_z$ . 

## The Phon formalism: Pitch

1. The eigenvectors of $\sigma_z$ are $\mid u \rangle$ and $\mid d \rangle$ (pitch up and pitch down):
   - $\sigma_z \mid u \rangle = \mid u \rangle $
   - $\sigma_z \mid d \rangle = - \mid d \rangle $
2. Orthogonality: $ \langle u \mid d \rangle = 0$

### Matrix representation
$\mid u \rangle = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$; $\mid d \rangle = \begin{bmatrix} 0 \\ 1 \end{bmatrix}; \sigma_z = \begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix} $

## The Phon formalism: Turbulence

1. The eigenvectors of $\sigma_x$ are $\mid l \rangle$ and $\mid r \rangle$ (noise left and noise right):
   - $\sigma_x \mid r \rangle = \mid r \rangle $
   - $\sigma_x \mid l \rangle = - \mid l \rangle $
2. Orthogonality: $ \langle l \mid r \rangle = 0$
3. If the phon is prepared $\mid r \rangle$, and then we measure $\sigma_z$, there will be equal
probabilities for $\mid u \rangle$ or $\mid d \rangle $ as an outcome: $\mid r \rangle = \frac{1}{\sqrt{2}} \mid u \rangle + \frac{1}{\sqrt{2}} \mid d \rangle$ (**Born rule**)
4. If the phon is prepared $\mid l \rangle$, and then we measure $\sigma_z$: $\mid l \rangle = \frac{1}{\sqrt{2}} \mid u \rangle - \frac{1}{\sqrt{2}} \mid d \rangle$

### Matrix representation
$\mid r \rangle = \begin{bmatrix} \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} \end{bmatrix}$; $\mid l \rangle = \begin{bmatrix} \frac{1}{\sqrt{2}} \\ -\frac{1}{\sqrt{2}} \end{bmatrix}; \sigma_x = \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix} $

## The Phon formalism: Slow Myoelastic Vibration

1. The eigenvectors of $\sigma_y$ are $\mid f \rangle$ and $\mid s \rangle$ (pulsation fast or slow):
   - $\sigma_y \mid f \rangle = \mid f \rangle $
   - $\sigma_y \mid s \rangle = - \mid s \rangle $
2. Orthogonality: $ \langle f \mid s \rangle = 0$
3. If the phon is prepared $\mid f \rangle$, and then we measure $\sigma_z$, there will be equal
probabilities for $\mid u \rangle$ or $\mid d \rangle $ as an outcome: $\mid f \rangle = \frac{1}{\sqrt{2}} \mid u \rangle + i \frac{1}{\sqrt{2}} \mid d \rangle$ (**Born rule**)
4. If the phon is prepared $\mid s \rangle$, and then we measure $\sigma_z$: $\mid s \rangle = \frac{1}{\sqrt{2}} \mid u \rangle - \frac{i}{\sqrt{2}} \mid d \rangle$

### Matrix representation
$\mid f \rangle = \begin{bmatrix} \frac{1}{\sqrt{2}} \\ \frac{i}{\sqrt{2}} \end{bmatrix}$; $\mid s \rangle = \begin{bmatrix} \frac{1}{\sqrt{2}} \\ -\frac{i}{\sqrt{2}} \end{bmatrix}; \sigma_y = \begin{bmatrix} 0 & -i \\ i & 0 \end{bmatrix} $

## Pauli matrices and the Bloch sphere

$\sigma_z = \begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix} $; $\sigma_x = \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix} $; $ \sigma_y = \begin{bmatrix} 0 & -i \\ i & 0 \end{bmatrix} $

### Measurement along an arbitrary direction

Orienting the measurement apparatus along an arbitrary direction $\overline{n} = \left[ n_x , n_y , n_z \right]'$ means taking a weighted mixture:
$$\sigma_n = \overline{\sigma}\cdot\overline{n} = \sigma_x n_x + \sigma_y n_y + \sigma_z n_z = \begin{bmatrix} n_z & n_x -i n_y \\ n_x + i n_y & -n_z \end{bmatrix} $$

## Harmonic plus Noise model

A measurement performed by means of a Harmonic plus Noise model would lie in the phonation-turbulence plane
($n_z = \cos\theta, n_x = \sin\theta, n_y = 0$), so that $\sigma_n =  \begin{bmatrix} \cos\theta & \sin\theta \\ \sin\theta & -\cos\theta \end{bmatrix} $


The eigenstate for eigenvalue $+1$ is $\mid\lambda_1\rangle = \begin{bmatrix}\cos{\theta/2} \\ \sin{\theta/2}\end{bmatrix}$, the eigenstate for eigenvalue $-1$ is $\mid\lambda_{-1}\rangle = \begin{bmatrix}-\sin{\theta/2} \\ \cos{\theta/2}\end{bmatrix}$, and $\langle\lambda_1\mid\lambda_{-1}\rangle = 0$.

Suppose we prepare the phon to pitch up $\mid u \rangle$. If we rotate the measurement system along $\overline{n}$,
the probability, the expectation value of the measurement is (by Born rule):
$$\langle \sigma_n \rangle = \sum_i \lambda_i P(\lambda_i) = \lambda_1 \mid \langle u \mid \lambda_1 \rangle \mid^2 + \lambda_{-1} \mid \langle u \mid \lambda_{-1} \rangle \mid^2 = (+1) \cos^2{\theta/2} + (−1) \sin^2{\theta/2} = \cos\theta$$

## Postulate 1 of Quantum Mechanics
At each time instant the system is completely specified by a state $\mid \psi \rangle$ such that $\langle \psi \mid \psi \rangle = 1$.
The state can be known with certainty or probabilistically as being one of a set of $\mid \psi_i \rangle$ with a given probability distribution (random or mixed state). [[G. Cariolaro, Quantum Communications, 2015](https://www.springer.com/it/book/9783319155999)]
### Density Operator
$$\rho = \sum_i p_i \mid \psi_i \rangle \langle \psi_i \mid $$

For a **pure state** $\rho = \mid \psi \rangle \langle \psi \mid $ and $Tr[\rho^2] = 1$.

For a **mixed state** $Tr[\rho^2] < 1$.

#### Example: 
State $\mid u \rangle $ with probability $\frac{1}{3}$ and state $\mid d \rangle$ with probability $\frac{2}{3}$.

$\rho = \frac{1}{3} \mid u \rangle \langle u \mid + \frac{2}{3} \mid d \rangle \langle d \mid = \begin{bmatrix} \frac{1}{3} & 0 \\ 0 & \frac{2}{3} \end{bmatrix}$

$Tr[\rho^2] = \frac{5}{9} < 1 $


## Uncertainty

If we measure two observables ${\bf L}$ and ${\bf M}$  simultaneously, the system is left
in a simultaneous eigenvector of the observables only if ${\bf L}$ and
${\bf M}$ commute, i.e. if their commutator $\left[ {\bf L, M} \right] = {\bf LM - ML}$ is null.  

Measurement operators along different axes do not commute. For example, $\left[
  \sigma_x, \sigma_y \right] = 2 i \sigma_z$, and therefore phonation
and turbulence can not be simultaneously measured with certainty.

### The uncertainty principle
\begin{equation} \label{uncertaintyPrinciple}
  \Delta {\bf L} \Delta {\bf M} \geq \frac{1}{2} \left| \langle{\psi | \left[{\bf L, M}\right] | \psi}\rangle \right|
\end{equation}


## Postulate 2 of Quantum Mechanics: Time evolution

Time evolution of states is described by Hamiltonian operator ${\bf H}(t)$, through Schrödinger equation $$\frac{\partial \mid \psi(t)\rangle}{\partial t} = - i {\bf H}(t) \mid \psi(t)\rangle $$ or, alternatively, by the unitary transformation
$$\mid \psi(t)\rangle = {\bf U}(t_0, t) \mid\psi(t_0)\rangle, t > t_0, $$ where the unitary operator ${\bf U} = {\bf U}(t_0, t)$ depends only on $t_0$ and $t$.

In case of **time independence** (${\bf H}(t) = {\bf H}$), the unitary operator is $${\bf U}(t_0, t) = {\bf U}(t - t_0) = e^{-i {\bf H} (t-t_0)}$$

Quantum evolution preserves the magnitude of states and the angle between states.

For non-pure states, the evolution of density operators is $$\rho(t) = {\bf U}^*(t_0, t) \rho(t_0) {\bf U}(t_0, t)$$

The system remains in a superposition of states until a measurement occurs, whose outcome is inherently probabilistic. <br>
The states change under the effect of external forces, which determine the change of probabilities. <br>
The Hamiltonian $$  {\bf H} = \frac{\omega}{2}  \overline{\sigma} \cdot \overline{n} = \frac{\omega}{2} \begin{bmatrix} n_z &  n_x - i n_y \\ n_x + i n_y & -n_z \end{bmatrix} $$ controls the evolution process. <br>
With a **commutative Hamiltonian** ($\left[{\bf H}(0),{\bf H}(t)\right] = 0 $), the time evolution can be expressed as
$$ \mid{\psi(t)}\rangle = e^{-i\int_0^t {\bf H}(\tau)d\tau}\mid{\psi(0)}\rangle.$$
In general, if $\left[{\bf A},{\bf B}\right] \neq 0$ we have that
$e^{\bf A} e^{\bf B} \neq e^{{\bf A}+{\bf B}}$. Since the evolution between two time points $0$ and $t$ can be split at an intermediate time $t^*$, if $e^{-i\int_0^t {\bf H}(\tau)d\tau} = e^{-i\int_0^{t^*} {\bf H}(\tau)d\tau  -i\int_{t^*}^t {\bf H}(\tau)d\tau} \neq e^{-i\int_0^{t^*} {\bf H}(\tau)d\tau} e^{ -i\int_{t^*}^t {\bf H}(\tau)d\tau}$ then it means that an explicit solution in terms of an integral can not be found.
Our approach is to consider time segments where the Hamiltonian is locally commutative, and to compute the time evolution segment by segment in terms of an integral.

### A commutative Hamiltonian that ensures convergence

$$ {\bf H}(t) = g(t) {\bf S}$$

with ${\bf S}$ a time-independent Hermitian matrix, and $g(t) = e^{-t}$. 

In image segmentation ([Youssry et al., 2015](https://link.springer.com/article/10.1007/s11128-015-1072-3)), time evolution leads to final foreground-background pixel assignment, and ${\bf S}$ depends on the neighborhood. 

In sound, in the context of signal separation, the Hamiltonian ${\bf H} = \frac{\omega}{2} \begin{bmatrix} n_z &  n_x - i n_y \\ n_x + i n_y & -n_z \end{bmatrix} $ can be made time-dependent, with the components of potential energy extracted as audio features on medium-time windows.

## Postulate 3 of Quantum Mechanics: Measurement
A projector system $\Pi_i$ in the Hilbert space is Hermitian, Idempotent, and Complete. 

If the system is in state $\mid \psi \rangle$ before measurement, the probability that the outcome of a measurement through a projector system returns $m = i$ is
$$ p_m(i|\psi)= P[m = i | s = \mid \psi \rangle] = \langle \psi \mid \Pi_i \mid \psi \rangle $$
and, as a result of the measurement, the system collapses in state 
$$\psi^{(i)}_{post} = \frac{\Pi_i \mid \psi \rangle }{\sqrt{p_m(i|\psi)}}$$

Given an orthonormal basis of **measurement vectors** $\mid a_i \rangle$, the elementary projectors are $\Pi_i = \mid a_i \rangle \langle a_i \mid $, 
$$ p_m(i|\psi) = |\langle \psi \mid a_i \rangle |^2 $$ and the system (by neglecting a unitary phasor) collapses into $$\psi^{(i)}_{post} = |a_i\rangle $$
If the system is in a pure state
    $$ p_m(i|\psi) = \langle \psi \mid \Pi_i \mid \psi \rangle = Tr[\rho \Pi_i]$$
If the system is in a mixed state, the outcome of measurement is formulated as a random variable conditioned by a given state:
$$ p_m(i|\psi_j) = P[m = i | s = \mid \psi_j \rangle] = \langle \psi_j \mid \Pi_i \mid \psi_j \rangle = Tr[\mid \psi_j \rangle \langle \psi_j \mid \Pi_i]$$
and by averaging over all components of the mixed state we get
$$P[m = i| \rho] = Tr[\rho \Pi_i] $$
with the system collapsing into the new ensemble of states represented by
$$\rho^{(i)}_{post} = \frac{\Pi_i \rho \Pi_i}{Tr[\rho \Pi_i] }$$

```python
# Generate a mixture of two crossing glissandos with a burst of lowpass-filtered noise. 
timeLength = 2.0 #seconds
fs = 44100
t = np.arange(0, timeLength, 1.0/fs)
ww = get_window("hann", t.size, 0.5)
noiseAmp = 0.8
noiseWidth = 0.2 #seconds
noiseLocation = 1.0 #seconds
wnoise = get_window(("tukey", 0.3), int(noiseWidth*fs))

Ar = .3
fi = 333.0 # initial frequency
ff = 133.0 # final frequency
phi0 = 0 
# linear frequency sweep
phi = phi0 + 2*np.pi*fi*t + np.pi*(ff - fi)/timeLength*t*t 
s1 = Ar * np.cos(phi)

fi = 200.0 # initial frequency
ff = 300.0 # final frequency
phi0 = 0 
# linear frequency sweep
phi = phi0 + 2*np.pi*fi*t + np.pi*(ff - fi)/timeLength*t*t 
s2 = Ar * np.cos(phi)

ss = (s1 + s2)
ss[int(noiseLocation*fs):(int(noiseLocation*fs)+int(noiseWidth*fs))] *= (1.0 - wnoise)
noise = np.zeros(phi.size) 
noiseInsert = np.random.normal(0, noiseAmp, int(noiseWidth*fs))
b, a = signal.butter(13, 0.1) # butterworth 13rd order lowpass
noiseInsert = signal.lfilter(b, a, noiseInsert) 
noise[int(noiseLocation*fs):(int(noiseLocation*fs)+int(noiseWidth*fs))] = \
     wnoise*noiseInsert

zs = (ss + noise)*ww
sd.play(zs, fs)

plt.figure(1, figsize=(15, 3))
plt.plot(t, zs, 'b', lw=1.5)
plt.axis([0, 2,-1.0,1.0])
plt.xlabel('time (seconds)')
plt.ylabel('amplitude')
plt.tight_layout()
plt.savefig('sine-synthesis.png')
plt.show()

Hop = 1024
M = 2048
N = 2*M
maxplotfreq = 4000.0
w = get_window('hann', M)
mX,pX = STFT.stftAnal(zs, w, N, Hop)
sizeEnv = int(mX[0,:].size)
binFreq = (.5*fs)*np.arange(sizeEnv*maxplotfreq/(.5*fs))/sizeEnv
numFrames = int(mX[:,0].size)
frmTime = Hop*np.arange(numFrames)/float(fs)                             
plt.pcolormesh(frmTime, binFreq, np.transpose(mX[:,:int(sizeEnv*maxplotfreq/(.5*fs)+1)]))
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.title('frequency crossing')

UF.wavwrite(zs, fs, "sines-noise-cross.wav")
```

```python
# Compute the sound features:
MAXINT = 32768
x = EqualLoudness()(x)
totalSaliences1 = []
totalSaliences2 = []
totalBins1 = []
totalBins2 = []
spectra = []
energyNoise1 = []
energyNoise2 = []
for frame in FrameGenerator(x, frameSize=M, hopSize=H):
    frame = window(frame)
    mX = spectrum(frame)
    mpX = fft(frame) 
    spectra.append(mX)
    peak_frequencies, peak_magnitudes = spectralPeaks(mX)  
    freqs, mags, phass = sineModelAnal(mpX) 
    deSined = sineSubtraction(frame, mags, freqs, phass) 
    deSinedSpectrum = spectrum(deSined)
    eb1 = EnergyBand(startCutoffFrequency=1000, stopCutoffFrequency=2000)(deSinedSpectrum)/MAXINT # for two bands
    eb2 = EnergyBand(startCutoffFrequency=2000, stopCutoffFrequency=6000)(deSinedSpectrum)/MAXINT
    energyNoise1.append(eb1)
    energyNoise2.append(eb2)
    pitchSalienceFunction_vals = pitchSalienceFunction(peak_frequencies, peak_magnitudes)
    salience_peaks_bins_vals, salience_peaks_saliences_vals = \
        pitchSalienceFunctionPeaks(pitchSalienceFunction_vals)
    # bins and vals are sorted
    if (size(salience_peaks_saliences_vals)>1): 
        totalSaliences1.append(salience_peaks_saliences_vals[0])
        totalSaliences2.append(salience_peaks_saliences_vals[1])
        totalBins1.append(salience_peaks_bins_vals[0])
        totalBins2.append(salience_peaks_bins_vals[1])
    else: # to deal with last frame singularity
        totalSaliences1.append(0)
        totalSaliences2.append(0)
        totalBins1.append(0)
        totalBins2.append(0)
totalSaliences1 = np.array(totalSaliences1)
totalSaliences2 = np.array(totalSaliences2)
totalBins1 = np.array(totalBins1)
totalBins2 = np.array(totalBins2)
totalFreqs1 = rF*pow(2,totalBins1*bR/1200) # cent bins must be transformed to Hz
totalFreqs2 = rF*pow(2,totalBins2*bR/1200) 
```

### Constructing a Hamiltonian
Compute the ${\bf S}$ matrix of the Hamiltonian from audio features

```python
decimation = 10 # PARAMETER (hold on): use integers larger than 1 to hold potential across frames
n_x = energyNoise1[::decimation]
n_x = np.repeat(n_x, decimation)[0:size(energyNoise1)] # turbulence potential
n_y = np.zeros(n_x.size) # n_y is not at play here
n_z = totalSaliences1[::decimation] 
n_z = np.repeat(n_z, decimation)[0:size(totalSaliences1)] # pitch potential
S = np.array([[n_z, n_x - 1j*n_y], [n_x + 1j*n_y, -n_z]])
```

Time-varying Hamiltonian multiplied by a decreasing exponential $g(m) = e^{- k m}$, where $m$ is the frame number, extending over $M_f$ frames:
<img src="./turbulence_pitchPotential1.png" width="400"> 

```python
k = 0.1 # PARAMETER (damping)
et = exp(-k*np.arange(0,decimation))
et = np.tile(et,int(size(n_z)/decimation))
ett = np.pad(et,(0,size(n_z)-size(et)),'constant',constant_values=(0))
nzc = n_z*ett
nxc = n_x*ett
nyc = n_y*ett
plt.subplot(2,1,1)
plt.title('Potentials')
plt.plot(nxc, color='r');
plt.plot(n_x, color='y');
plt.ylabel('turbulence potential')
plt.subplot(2,1,2)
plt.plot(nzc, color='r');
plt.plot(n_z, color='y');
plt.ylabel('pitch potential')
plt.xlabel('frame number')
savefig('turbulence_pitchPotential1.pdf')
```

### Integrate and exponentiate the Hamiltonian
approximate $ e^{-i\int_0^t H(\tau)d\tau} $

```python
H = np.array([[nzc, nxc - 1j*nyc], [nxc + 1j*nyc, -nzc]]) 
intHc = np.cumsum(H, axis=2) # integral of H over all time (cumulative sum)
Uc = intHc
for t in range(size(n_z)):
    Uc[:,:,t] = linalg.expm(-(1j)*intHc[:,:,t])  
```

## Example: Evolution from pitch-up
<audio src="evolutionUp.wav" controls preload></audio> 
<img src="evolutionUp.png" width="600"> 

```python
# given the initial state "pitch up", compute the state at all future times
s0 = np.matrix([[1], [0]])
sT = np.zeros((2, size(n_z)), dtype=complex)
for t in range(size(n_z)):
    sT[:,[t]] = Uc[:,:,t] * s0
```

```python
threshold = 0.7 # PARAMETER (pitchiness threshold)
hopCollapse = 5 # PARAMETER (decimation of collapses)
sigma_z = [[1, 0], [0, -1]]
sigma_x = [[0, 1], [1, 0]]
traccia = zeros(size(n_z))
sbuffi1 = zeros(size(n_z))
sbuffi2 = zeros(size(n_z))
for t in range(size(totalFreqs1)): # select pitch up or pitch down based on state
    sdiff = norm(transpose(matrix(sT[:, t])) - sigma_z * transpose(matrix(sT[:, t])))
    if (sdiff < threshold or sdiff > (2 - threshold)): # state is pitchy (how close to eigenvalue of sigma_z)
        prob = np.square(abs(sT[0,t])) # collapse based on probability
        cstate = np.random.choice(np.arange(0, 2), p=[prob, 1-prob]) 
        if (cstate == 0):
            if (t%hopCollapse == 0):
                sT[0,t] = 1.0
                sT[1,t] = 0.0
            traccia[t] = totalFreqs1[t] if (totalFreqs1[t] >= totalFreqs2[t]) else totalFreqs2[t]
        else: 
            if (t%hopCollapse == 0):
                sT[0,t] = 0.0
                sT[1,t] = 1.0 
            traccia[t] = totalFreqs1[t] if (totalFreqs1[t] < totalFreqs2[t]) else totalFreqs2[t]
    else: # state is turbulent
        prob = abs(np.square(np.dot(sT[:,t],[1/sqrt(2),1/sqrt(2)]))) # collapse based on probability
        cstate = np.random.choice(np.arange(0, 2), p=[prob, 1-prob]) 
        if (cstate == 0):
            if (t%hopCollapse == 0):
                sT[0,t] = 1/sqrt(2)
                sT[1,t] = 1/sqrt(2)
            sbuffi1[t] = energyNoise1[t]
        else:
            if (t%hopCollapse == 0):
                sT[0,t] = 1/sqrt(2)
                sT[1,t] = -1/sqrt(2)
            sbuffi2[t] = energyNoise2[t]
    # recompute future from time t onward
    if (t%hopCollapse == 0):
        intHc[:,:,t:] = np.cumsum(H[:,:,t:], axis=2)
        for tt in range(t,size(n_z)):
            Uc[:,:,tt] = linalg.expm(-(1j)*intHc[:,:,tt])
            sT[:,[tt]] = Uc[:,:,tt] * matrix(sT[:,[t]])
```

```python
plt.figure(2, figsize=(9.5, 7))
plt.subplot(4,1,1)
plt.title('Hamiltonian evolution from pitch up')
plot(frmTime, totalFreqs1[0:size(frmTime)], color='r', linewidth = 1.5)
plot(frmTime, totalFreqs2[0:size(frmTime)], color='y', linewidth = 1.5)
scatter(frmTime, traccia[0:size(frmTime)], color='g', linewidth = 0.1)
plt.axis([0, x.size/float(fs), 0, max(totalFreqs1)*2])
plt.ylabel('pitch')
plt.xlabel('Time[sec]')
lineTraccia = np.interp(tfull,tsparse,traccia[0:size(tsparse)])
tsSbuffi1 = np.interp(tfull,tsparse,sbuffi1[0:size(tsparse)])
tsSbuffi2 = np.interp(tfull,tsparse,sbuffi2[0:size(tsparse)])
plt.subplot(4,1,2)
plot(ts1+ts2)
sbuffis1 = tsSbuffi1*(rand(size(line1)) - 0.5)  
sbuffis1 = BandPass(cutoffFrequency = 1500,bandwidth=1000)(single(sbuffis1))
if max(sbuffis1) > 0.001 :
    sbuffis1 = sbuffis1/max(sbuffis1)*0.5
sbuffis2 = tsSbuffi2*(rand(size(line1)) - 0.5)                            
sbuffis2 = BandPass(cutoffFrequency = 4000,bandwidth=1500)(single(sbuffis2))
if max(sbuffis2) > 0.001 :
    sbuffis2 = sbuffis2/max(sbuffis2)*0.5
phase = np.cumsum(lineTraccia)
lineTraccias = (ts1+ts2)*sin(2*np.pi*phase/fs) 
spn = lineTraccias  + sbuffis1 + sbuffis2
plt.subplot(4,1,3)
plot(sbuffis1)
plt.subplot(4,1,4)
plot(sbuffis2)
savefig('evolutionUp.png')
sd.play(spn, fs)
UF.wavwrite(spn, fs, "evolutionUp.wav")
```

## Evolution from mixed state
Consider the two crossing glides interrupted by noise, but this time the initial state is known only probabilistically. For example, consider the initial mixture of $\frac{1}{2}$ pitch-up and $\frac{2}{3}$ pitch-down. 

The density matrix $$\rho = \begin{bmatrix} \frac{1}{3} & 0 \\ 0 & \frac{2}{3} \end{bmatrix}$$
is evolved according to $$\rho(t) = {\bf U}^*(t_0, t) \rho(t_0) {\bf U}(t_0, t),$$ where the unitary operator is defined as $${\bf U}(t_0,t) = e^{-i\int_0^t {\bf H}(\tau)d\tau}.$$
When a pitch measurement is taken, the outcome is up or down according to 
$$P[m = i| \rho] = Tr[\rho \Pi_i], $$ and the density matrix that results from collapsing upon measurement is given by 
$$\rho^{(i)}_{post} = \frac{\Pi_i \rho \Pi_i}{Tr[\rho \Pi_i] }.$$

The density matrix can be made audible in various ways, thus sonifying the Hamiltonian evolution. For example, the completely chaotic mixed state, corresponding to the half-identity matrix $\rho = \frac{1}{2} {\bf I}$, can be made to sound as noise, and the pure states can be made to sound as the upper or the lower of the most salient pitches. These three components can be mixed for intermediate states. If $p_u$ and $p_d$ are the respective probabilities of pitch-up and pitch-down as encoded in the mixed state, the resulting mixed sound can be composed by a noise having amplitude $\min{(p_u, p_d)}$, by the upper pitch weighted by $p_u - \min{(p_u, p_d)}$, and by the lower pitch weighted by $p_d - \min{(p_u, p_d)}$.

## Example: Evolution from pitch-up

```python
H = np.array([[nzc, nxc - 1j*nyc], [nxc + 1j*nyc, -nzc]]) 
intHc = np.cumsum(H, axis=2) # integral of H over all time (cumulative sum)
Uc = intHc                   # compute the unitary operator U
for t in range(size(n_z)):
    Uc[:,:,t] = linalg.expm(-(1j)*intHc[:,:,t])  
rho = np.matrix([[1/3, 0], [0, 2/3]]) # the initial mixed state 
# compute the density matrix at all future times
rhoT = np.zeros((2, 2, size(n_z)), dtype=complex)
for t in range(size(n_z)):
    rhoT[:,:,t] = (np.matrix(Uc[:,:,t])).getH() * rho * Uc[:,:,t]
```

```python
threshold = 0.7 # PARAMETER (pitchiness threshold)
hopCollapse = 5 # PARAMETER (decimation of collapses)
pi_up = [[1, 0], [0, 0]]
pi_down = [[0, 0], [0, 1]]
amplitude_up = zeros(size(n_z))
amplitude_down = zeros(size(n_z))
amplitude_noise = zeros(size(n_z))
sbuffi1 = zeros(size(n_z))
sbuffi2 = zeros(size(n_z))
for t in range(size(totalFreqs1)): 
    p_up = abs((rhoT[:,:,t] * pi_up).trace())
    p_down = abs((rhoT[:,:,t] * pi_down).trace())
    amplitude_noise[t] = min(p_up, p_down)
    amplitude_up[t] = p_up - amplitude_noise[t]
    amplitude_down[t] = p_down - amplitude_noise[t]
    cstate = np.random.choice(np.arange(0, 2), p=[p_up, p_down]) 
    if (cstate == 0):
        if (t%hopCollapse == 0):
            rhoT[:,:,t] = pi_up * rhoT[:,:,t] * pi_up / (rhoT[:,:,t] * pi_up).trace()
    else:
        if (t%hopCollapse == 0):
            rhoT[:,:,t] = pi_down * rhoT[:,:,t] * pi_down / (rhoT[:,:,t] * pi_down).trace()
    # recompute future from time t onward
    if (t%hopCollapse == 0):
        intHc = np.cumsum(H, axis=2) 
        Uc = intHc                   # compute the unitary operator U
        for tt in range(t,size(n_z)):
            Uc[:,:,tt] = linalg.expm(-(1j)*intHc[:,:,tt])  
            rhoT[:,:,tt] = (np.matrix(Uc[:,:,tt])).getH() * rhoT[:,:,t] * Uc[:,:,tt]
```

```python
plt.figure(4)
plt.title('Hamiltonian evolution from mixed state')
plot(frmTime, amplitude_up, color='r', linewidth=1.5)
plot(frmTime, amplitude_down, color='y', linewidth=1.5)
plot(frmTime, amplitude_noise, color='g', linewidth=1.5)
legend(["pitch-up", "pitch_down", "noise"])
plt.axis([0, x.size/float(fs), 0, 1.05])
plt.ylabel('amplitude')
plt.xlabel('Time[sec]')
savefig('evolutionMixed.pdf')
```
